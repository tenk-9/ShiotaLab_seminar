## 対応リスト作成
照合時のペアリストを作成する．

`CS_seminar/_data/`に`wav_pair_list.txt`が生成されます．
この対応リストを前回の.ipynbに適用すれば話者照合できます．

```python
%cd ../
```

```python
import re
import itertools

def is_same_speaker(path1, path2):
  dir1 = re.match(r"(.*/).+\.wav", path1).group(1)
  dir2 = re.match(r"(.*/).+\.wav", path2).group(1)
  return int(dir1 == dir2)

same = 0

pairs = list(itertools.combinations(wavs_path, 2))
print(f"pairs:{len(pairs)}")

with open("wav_pair_list.txt", "w") as outfile:
  for pair in pairs:
    same_speaked = is_same_speaker(*pair)
    if same_speaked:
      same += 1
    col = f"{same_speaked} {pair[0]} {pair[1]}\n"
    outfile.write(col)

print(f"true pair:{same}\nfalse pair{len(pairs)-same}")
```
