```python
# soxを使ってノイズを重畳しましょう。

# ホワイトノイズの重畳
!sox ./NF001001_HS.wav -p synth whitenoise vol 0.02 | sox -m ./NF001001_HS.wav - ./addednoise.wav

# "synth whitenoise" でホワイトノイズを重畳することを指定、vol 0.02 でどのくらいの音量のノイズを重畳するかを指定します。
# オプションで重畳できるノイズはsine（正弦波）,square（矩形波）,triangle（三角波）,sawtooth（のこぎり波）,trapezium（台形波）,noise/whitenoise（ホワイトノイズ）
# tpdfnoise（三角形確率密度関数ノイズ）,pinknoise（ピンクノイズ）,brownnoise（ブラウニアンノイズ）,pluck（撥弦） が選択できるそうです。

# "vol 0.02" で加えるノイズの大きさを調整します。0~1の値を指定できます。ここの数値を大きくしすぎると、クリッピング(音割れ)が発生するので注意。

# "./NF001001_HS.wav"はノイズを加えたい音声ファイルのパス、"./addednoise.wav"はノイズを加えた後の音声を出力するためのパスに適宜変更してください。
```

```python
import librosa
import IPython

# 加工前の音声の確認
x, sr = librosa.load("/content/NF001001_HS.wav")
display(IPython.display.Audio(x, rate=sr))
```

```python
# ノイズを加えた音声の確認
y, sr = librosa.load("/content/addednoise.wav")
display(IPython.display.Audio(y, rate=sr))
```
